```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
# Funcional
plt.plot(x, y, color='r') # ou tirar o color e adicionar apenas 'r--'
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('TÃ­tulo')
```

```python
plt.subplots(nrows=1, ncols=2)
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x, y)
axes1.set_xlabel('Eixo X')
axes1.set_ylabel('Eixo Y')
axes1.set_title('TÃ­tulo')
axes2.plot(y, x)
```

PARTE 2

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g-.')
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Titulo do Grafico')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
ax
```

```python
for axe in ax:
    axe.plot(x, y, 'b')
    axe.set_title('Titulo')
fig
```

```python
ax[0].plot(x, y, 'b')
ax[0].set_title('Titulo')

ax[1].plot(x, y, 'b')
ax[1].set_title('Titulo')

fig
```
